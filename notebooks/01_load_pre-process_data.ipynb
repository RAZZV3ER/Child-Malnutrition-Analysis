**ABSTRACT:**
This research investigates child malnutrition trends across countries using the UNICEF Joint Child Malnutrition Estimates dataset. Key indicators such as stunting, wasting, overweight, and underweight among children under five years of age are analyzed over time and across regions. The study aims to identify patterns, regional disparities, and progress toward Sustainable Development Goals related to child health and nutrition. Data cleaning, visualization, and basic statistical analysis are employed to extract insights, providing a clear understanding of global malnutrition challenges and highlighting areas requiring policy attention.

```python
import os
import pandas as pd

# Define the file path
file_path = os.path.join("..", "data", 'raw', 'main', "JME_Country_Estimates_April_2021.xlsx")

# Load the sheet names
xls = pd.ExcelFile(file_path)
print("Sheets available:\n", xls.sheet_names, "\n")

# Preview the first 5 rows of each sheet
for sheet in xls.sheet_names:
    print(f"--- Preview of sheet: {sheet} ---")
    df = pd.read_excel(file_path, sheet_name=sheet)
    print(df.head(5))
    print("\n")
```

From this preview, the best approach for this research would be using the 'Survey Estimates' sheet, as it seems to contain the country-wise malnutrition data...

```python
# Load the 'Survey Estimates' sheet
df = pd.read_excel(file_path, sheet_name='Survey Estimates')

# Quick look
print(df.shape)
df.head()
```

```python
# Drop fully empty columns
df = df.dropna(axis=1, how='all')

# Drop fully empty rows
df = df.dropna(axis=0, how='all')

# Check the cleaned head
df.head()
print(df.shape)
```

```python
# Make a copy first
df_clean = df.copy()

# Clean column names
df_clean.columns = df_clean.columns.str.strip()  # Remove spaces at start/end
df_clean.columns = df_clean.columns.str.replace('*', '', regex=False)  # Remove asterisks
df_clean.columns = df_clean.columns.str.replace(' ', '_')  # Replace spaces with underscores
df_clean.columns = df_clean.columns.str.replace('-', '_')  # Replace dashes with underscores
df_clean.columns = df_clean.columns.str.replace("'", '', regex=False)  # Remove single quotes
df_clean.columns = df_clean.columns.str.replace("(", '', regex=False)  # Remove single quotes
df_clean.columns = df_clean.columns.str.replace(")", '', regex=False)  # Remove single quotes

# Check cleaned column names
df_clean.columns
```

```python
#Check for missing values
df_clean.isnull().sum()
```

```python
#Selecting Relevant Columns
cols_to_keep = [
    'ISO_code', 'Country_and_areas', 'Year', 'SDG_Region', 'UNICEF_Region', 
    'WHO_Region', 'Wasting', 'Stunting', 'Overweight', 'Underweight', 'U5_Population_000s'
]

df_analysis = df_clean[cols_to_keep].copy()
```

```python
#Check missing values in the new subset dataframe

print(df_analysis.isnull().sum())
```

```python
#Handling Missing Values

#For numeric columns, we can fill missing values with the mean or median
numeric_cols = ['Wasting', 'Stunting', 'Overweight', 'Underweight', 'U5_Population_000s']
for col in numeric_cols:
    df_analysis[col] = df_analysis[col].fillna(df_analysis[col].median())


#For categorical columns, we can fill missing values with 'Unknown' or the mode
categorical_cols = ['UNICEF_Region', 'WHO_Region']
for col in categorical_cols:
    df_analysis[col] = df_analysis[col].fillna('Unknown')

#Verify no missing values remain
print(df_analysis.isnull().sum())
```

```python
# Define path to save cleaned dataset
processed_path = os.path.join("..", "data", "processed_and_cleaned", "child_malnutrition_cleaned.csv")

# Save the cleaned dataframe
df_analysis.to_csv(processed_path, index=False)

print(f"Cleaned dataset saved to: {processed_path}")
```
